```python
def get_country_code(country_name): #defini un méthode pour trouver obtenir le code de chaque pays
    for code, name in COUNTRIES.items():
        if name == country_name:
            return code
    return None
def get_value(str_money):   #parce que dans le .csv les donnes de "value" ne sont pas des chiffres, c'est un changement de chiffre
    str_money=str_money[1:]
    str_money=str_money[:-1] 
    return str_money


def str2float(s): #c'est pour changer string à float
    def fn(x,y):
        return x*10+y
    n=s.index('.')
    s1=list(map(int,[x for x in s[:n]]))
    s2=list(map(int,[x for x in s[n+1:]]))
    return functools.reduce(fn,s1)+functools.reduce(fn,s2)/(10**len(s2))

dict_0 = {} #stocker le nombre de joueurs de chaque pays
dict_total={} #stocker le values totales de joueurs de chaque pays
dict_mo= {}#stocker les valuse moyens joueurs de chaque pays
with open('data.csv',encoding='utf-8', errors='ignore',newline='') as csvfile:
     reader = csv.DictReader(csvfile)
     for row in reader:
         na = row['Nationality']         
         if(row['Nationality']in dict_0):
            cnt = dict_0.get(row['Nationality'])
            #  dict.update({'nationality':na,'count':cnt+1})
            dict_0[na]=cnt+1
            try:
                dict_total[na]+=float(get_value(row['Value']))
            except:
                dict_total[na]+=0
         else:
            #dict.update({'nationality':na,'count':1})
            dict_0[na]=1
            try:
                dict_total[na]=float(get_value(row['Value']))
            except:
                dict_total[na]=0
for key,value in dict_0.items():
    if(key!='India'):
        dict_mo[key]=dict_total[key]/dict_0[key]#c'est pour calculer le moyens
str1='{'
for key, value in dict_mo.items():
    if(str(get_country_code(key))!='None'):
        str1+='\''+str(get_country_code(key))+'\''+':'+str(value)+','
str1=str1[:-1]              

str1+='}'
str1=eval(str1)#c'est pour obtenir le String utilisé d'après
#print(str1)

%matplotlib inline

html_pygal = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
    <!-- ... -->
  </head>
  <body>
    <figure>
      {pygal_render}
    </figure>
  </body>
</html>
"""
import pygal.maps.world
from pygal.style import RedBlueStyle as RB
worldmap_chart = pygal.maps.world.World(style=RB)
worldmap_chart.title = "average value of players in each region"
worldmap_chart.add('average value',str1)
worldmap_chart.render_to_file('./2.svg')
HTML(html_pygal.format(pygal_render=worldmap_chart.render()))
```
